## FIND OUT ABOUT DataFrame.Pivot_table method

```python
q5set2015 = dataset2015.loc[:, ["Country", "Happiness Score"]]
q5set2017 = dataset2017.loc[:, ["Country", "Happiness.Score"]]
```

```python
mer = pd.merge(left=q5set2015, right=q5set2017, on='Country')
```

```python
mer[["HS 2015", "HS 2017"]] = mer[["Happiness Score", "Happiness.Score"]]
```

```python
mer =mer.drop(["Happiness Score", "Happiness.Score"], axis=1)
```

```python
mer[mer["HS 2015"] - mer["HS 2017"] > 0]
```
